Feature engineering

```python
df=pd.read_csv('my clean.csv')
```

```python
# we can remove Is delivering now column as majority 98-99 percent values are same
```

```python
df.drop(columns={'Is delivering now'},inplace=True)
```

```python
# we will encode has table booking and has online delivery
```

```python
def transf(x):
  if x=='Yes':
    return 1
  else: return 0

df['Has Online delivery']=df['Has Online delivery'].apply(transf)
df['Has Table booking']=df['Has Table booking'].apply(transf)
```

```python
# we will create a new row which will have all the average cost for two
# values in indian rupees
```

```python
currency_to_inr = {
    'Indian Rupees': 1,
    'Dollar': 83,
    'Pounds': 105,
    'Qatari Rial': 22.8,
    'Brazilian Real': 16,
    'Rand': 4.5,
    'Emirati Dirham': 22.6,
    'New Zealand Dollar': 51,
    'Turkish Lira': 2.6,
    'Botswana Pula': 6.1,
    'Indonesian Rupiah': 0.0052,
    'Sri Lankan Rupee': 0.25

}

df['avg cost in inr']=df['Currency'].apply(lambda x:currency_to_inr[x])*df['Average Cost for two']
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.drop(idx,inplace=True)
```
